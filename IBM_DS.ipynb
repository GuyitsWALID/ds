```python
import requests
import pandas as pd
import io

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
urlData = requests.get(url).content
df = pd.read_csv(io.StringIO(urlData.decode('utf-8')), header = None)
```

```python
df.head()
```

```python
df.columns = headers
df.head()
```

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
df["bore"] = df["bore"].replace(np.nan, avg_bore, inplace=True)
average_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke: ", average_stroke)
df["stroke"] = df["stroke"].replace(np.nan, average_stroke, inplace =True)
df.head(20)
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)


```

```python
df['horsepower'] = df['horsepower'].replace(np.nan, avg_horsepower)
```
